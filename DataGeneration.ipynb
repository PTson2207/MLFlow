```python
import numpy as np
import pandas as pd
from matplotlib import  pyplot as plt
from sklearn.datasets import  make_blobs
import seaborn as sns
import random
```

```python
sns.set_theme()
```

```python
def data_generator(name):
    X, y = make_blobs(n_samples=1000, center_box=(-10.0, 10.0), cluster_std=2, centers=2)
    print(X.shape)
    print(y.shape)

    data = {"Coord_X": X[:, 0], "Coord_Y": X[:, 1], "label": y}
    df = pd.DataFrame.from_dict(data)

    df["label"] = df.label.replace(0, "Blue")
    df["label"] = df.label.replace(1, "Red")

    plt.figure(figsize=(10, 10))
    ax = sns.scatterplot(data=df, x=df.Coord_X, y=df.Coord_Y, hue=df.label)
    ax.set_title(name)
    plt.show()

    df.to_csv("./data/" +name+ ".csv", index=False)
    
```

```python
dataset = ["sample" + str(i) for i in range(1, 11)]
for data in dataset:
    data_generator(data)
```

```python
dataset = ["sample"+str(i) for i in range(1,11)]
figure, axes = plt.subplots(5, 2)
figure.set_figwidth(20)
figure.set_figheight(26)
figure.subplots_adjust(hspace=0.3)
plt.title("Data Samples")
x = 0
y = 0
for i, data in enumerate(dataset):
    df = pd.read_csv("./data/"+data+".csv")
    ax = sns.scatterplot(data=df,x=df.Coord_X,y=df.Coord_Y,hue=df.label,ax=axes[x,y])
    ax.set_title(data)
    
    if y == 0:
        y = 1
    else:
        y = 0
        x = x +1
plt.show()
figure.savefig("data_samples.jpg",dpi=300)
```
